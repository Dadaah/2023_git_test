## 함수 만들기
### 함수
* 함수는 입력값을 받아 출력값을 반환하는 코드의 묶음
* 기능을 반복해서 구현할 필요가 없으므로 코드를 간결하게 유지할 수 있다.
#### 정의하는 법
* def와 함수이름 그리고 함수의 본문을 이용하여 정의한다.
* return을 이용해서 반환한다.
### 입력과 반환값에 따른 함수구조 

#### 1. 입력도 없고 반환값도 없는 함수

* 구조

def 함수명() :
   수행문장
   ...
* 예시

```python
def func1():
    print('BlockDMask')
```
반환값은 없지만 함수를 실행하면 'BlockDMask'를 프린트한다.
#### 2. 입력만 있는 함수

* 구조

def 함수이름(매개변수1, 매개변수2, ... ):
    수행문장

* 예시

```python
def func2(a, b):
    print(f'{a} 곱하기 {b} = {a * b}')
func2(1, 2)
func2(1, 3)
func2(2, 4)
```
반환값은 없다 그러면 이제 반환값이 있는 함수를 알아보자
#### 3. 반환값만 있는 함수

* 구조

def 함수이름():
    수행문장
    return 반환값

* 예시

```python
def func3():
    return "abcdefg"
a = func3()
print(a + "GG")
```
a에 "abcdefg"이 반환된다.

#### 4. 입력, 반환값이 둘다 있는 함수(가장 흔한 case)
가장 많이 쓰이는 case이다. 

* 구조

def 함수이름(매개변수1, 매개변수2 ...):
    수행문장
    return 반환값

* 예시

```python 
def func4(a, b):
    return a * b
    c = func4(3, 9)
    print(c)
```
3+9의 값이 c에 반환된다.
### 파이썬 함수 예제
##### 구구단 함수 만들기
 * def를 사용하지 않고 for문으로 만들기
```python
    for i in range(1, 10):
        print(f'{2} x {i} = {2 * i}')
    for i in range(1, 10):
        print(f'{3} x {i} = {3 * i}') 
    # ...
    for i in range(1, 10):
        print(f'{9} x {i} = {9 * i}')
* def 함수를 사용할때
```python
def gugudan(num):
    for i in range(1, 10):
        print(f'{num} x {i} = {num * i}')
# 구구단 출력
gugudan(2)
gugudan(3)
# ...
gugudan(9)

```
이처럼 똑같이 행동하는 로직을 함수로 묶어버리면 코드가 훨씬 깔끔해 질 수 있고, 같은 코드를 작성하지 않아도 된다.
