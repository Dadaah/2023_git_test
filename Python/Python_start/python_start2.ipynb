 ### 리스트 
1. #### 리스트 선언  
`list_name=[] or list_name=list() or list_name=["abc",12,type]` 
list안에는 여러타입을 넣을 수 있다  
2. #### 리스트 추가  
`list_name.append(데이터) `  
`list.append("a")`  
` print(list)`  
`>>> ["abc",12,type,"a"]` 
마지막에 들어간다  
만약 마지막이 아니라 중간에 넣고 싶다면?  
`list.insert(들어갈인덱스번호진짜번호만 , 데이터)`
 0은 첫번째 -1은 뒤에서 2번째 마지막은 전체개수-1 이다  
<주의>list_name.append([4,5]) 이렇게 하면 두개의 데이터가 들어가는게 아닌 [4,5]가 하나의 데이터로 들어간다    
두개 이상을 넣고 싶다면?   
`list.extend(리스트)` 를 넣으면 된다 가장 마지막에 들어가고 한개를 넣더라도 []형태로 넣어야한다
3. #### 리스트 삭제  
`list_name.remove(삭제할데이터)` or  
`del list_name[인덱스번호]`  
4. #### 리스트 수정  
`list_name[인덱스번호]=바꿀데이터`
5. #### 리스트 정렬
`list_name.sort()` 유니코드 기반으로 정렬 [1,2,3,4 ... a,b,c ....]   
`list.reverse()` [4,3,2,1 ....]    
### tuple () 
1. 선언 및 입력
리스트와 비슷하다  
`tuple_name=() or (1,2,3,4)  tuple_name=tuple()`  
2. 읽기 변수명  
리스트와 비슷하다  
`tuple_name[인덱스번호]`    
3. 추가삭제수정이 불가능하다  
4. tuple끼리 더하기 반복은 가능    
`dat1=(1,2,3) dat2=(4,5,6)`  
`dat1+dat2 (1,2,3,4,5,6)`  
`dat1 x3   (1,2,3,1,2,3,1,2,3)`  
5. tuple의 꿀팁(변수끼리 값 바꾸기)  
    *원래   
새로운변수(temp)를 만들어서    
`temp=x #x를 temp에 넣어놓고`  
`x=y   #y를 x에 덮어놓고`  
`y=temp #temp에 덮어논 x를 y에 덮기`  
    *tuple 활용  
`(x,y)=(y,x)`  
6. list<->tuple    
list(tuple) tuple(list) 서로 변환이 가능하다
